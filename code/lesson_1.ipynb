```python
total_emissions = 0 

for point in country_data:
    total_emissions += point["mtCO2"]  #takes the value from the left part and adds the value from the right part

mean_emissions = total_emissions / len(country_data)
mean_emissions
```

```python
more_than_average_countries = []

for point in country_data:
    if point["mtCO2"] > mean_emissions:
        more_than_average_countries.append(point) #if the condition is True point is appended to the list N.B. mtCO2 is also a list

len(more_than_average_countries)
```

### start with the list. process each of its elements. append the result of our processing to some other list. we can use list comprehension instead of the code above to create more than average countries list. simplified syntax when we need to process elements of a list or some other collection one by one

```python
more_than_average_countries = [point for point in country_data if point["mtCO2"] > mean_emissions]

len(more_than_average_countries) #point is every point that specifies the condition
```

### another collection is a set(). it is unordered collection where each element can be present at most once. repeated elements in a set are deleted

```python
scandinavian_countries = set() #creates an empty set using set() function
print(scandinavian_countries)

scandinavian_countries.add("Sweden") #adds an element to a set
print(scandinavian_countries)

scandinavian_countries.add("Sweden") #if same element is added nothing changes since each element can be present at most once
print(scandinavian_countries)

scandinavian_countries.update(['Sweden', 'Denmark','Finland','Norway','Iceland']) #adds the collection of elements to the set
print(scandinavian_countries)

northern_europe = set(['Sweden', 'Denmark','Finland','Norway','Iceland','Netherlands','Luxembourg','Belgium']) #set() creates sets from other collection types 
print(northern_europe) 
```

```python
print('scandinavian_countries =', scandinavian_countries)
print('northern_europe =', northern_europe)

print('scandinavian_countries & northern_europe = {}'.format(scandinavian_countries & northern_europe))
print('scandinavian_countries | northern_europe = {}'.format(scandinavian_countries | northern_europe))

print('scandinavian_countries - northern_europe = {}'.format(scandinavian_countries - northern_europe))
print('northern_europe - scandinavian_countries = {}'.format(northern_europe - scandinavian_countries))

print('scandinavian_countries ^ northern_europe = {}'.format(scandinavian_countries ^ northern_europe))

print('scandinavian_countries >= northern_europe = {}'.format(scandinavian_countries >= northern_europe))
print('scandinavian_countries <= northern_europe = {}'.format(scandinavian_countries <= northern_europe))
print('scandinavian_countries == northern_europe = {}'.format(scandinavian_countries == northern_europe))
```
