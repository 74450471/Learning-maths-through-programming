<h3>Limit all lines to a maximum of 79 characters</h3>

```python
# WRONG
message = "This is a very long message that has been written all on one single line. Very-very-very long."
print(message)
print(len(message))
print(type(message))
```

```python
# Right
message = (
    "This is a very long message that has been split "
    "across multiple lines. Very-very-very long."
)
print(message)
print(len(message))
print(type(message))
```

<h3>Avoid extra space in some situations:</h3>

For example, immediately inside parentheses, brackets or braces:

```python
# WRONG
ham = ['ham 1', 'ham 2', 'ham 3']
eggs = 10

def spam(value1, value2):
    print(value1, value2)
    
spam( ham[ 1 ], { eggs: 2 } )
```

```python
# Right
spam(ham[1], {eggs: 2})
```

or immediately before the open parenthesis that starts the argument list of a function call:

```python
# WRONG
def print_some_text(text):
    print(text)
    
print_some_text ('My text') # it shoudn't be any extra spaces between function parenthesis
```

```python
# Right
def print_some_text(text):
    print(text)
    
print_some_text('My text') # print_some_text() - here we call a function, and ('My text') is argument of function
```

Compound statements (multiple statements on the same line) are generally discouraged:

```python
# WRONG

if foo == 'blah': do_blah_thing()
do_one(); do_two(); do_three()
```

```python
# Correct:
if foo == 'blah':
    do_blah_thing()
do_one()
do_two()
do_three()
```

<h3>1. Numerical operations</h3>

You can find cheatsheet on numerical operations for Python here: https://cheatography.com/mariofreitas/cheat-sheets/python-basics-math/

```python
x1 = 10
x2 = 20

print(x1 + x2) # addition
print(x1 - x2) # subtraction
print(x1 * x2) # multiplication
print(x1 / x2) # divison
print(x1 % x2) # get the remainder of the divison of x1 on x2
print(x1 // x2) # division without remainder
print(x1 ** x2) # raising a number to a power
```

<h3>2. Creating variables</h3>

In Python, we have some basic data types:
* int - integer numbers;
* float - float numbers;
* str - strings;
* bool - boolean values;

```python
x1 = 3
print(f'Type of x1 (value {x1}) = {type(x1)}') # integer - integer numbers
x2 = 3.01
print(f'Type of x2 (value {x2}) = {type(x2)}') # numbers with float
x3 = 'some text'
print(f'Type of x3 (value {x3}) = {type(x3)}') # str - string
x4 = False
print(f'Type of x4 (value {x4}) = {type(x4)}') # bool - boolean
```

And we have a few types of **collections** - the type of data that contains something inside:
* list - lists;
* dict - dictionaries;
* set - sets;
* tuples - tuples.

```python
x4 = [1, 2, 3, 4, 5]
x5 = (1, 2, 3, 4, 5)
x6 = {1: 'one', 2: 'two', 3: 'three'}
x7 = {1, 2, 3, 4, 5}

print(f'Type of x4 (value {x4}) = {type(x4)}') # list
print(f'Type of x5 (value {x5}) = {type(x5)}') # tuple
print(f'Type of x6 (value {x6}) = {type(x6)}') # dictionary
print(f'Type of x6 (value {x7}) = {type(x7)}') # set
```

<h3>3. Logical operations</h3>

* `and` – Returns True if both operands are True.
* `or` – Returns True if at least one operand is True.
* `not` – Returns True if the operand is False.

Read more about it here: https://www.geeksforgeeks.org/python/python-logical-operators/

```python
first = True
second = False
print(first and second) # True and False -> False
print(first or second) # True or False -> True
print(first and not(second)) # True and not(False) -> True
```


* `==` – Returns True if operands are equal.
* `!=` – Returns True if operands are not equal.
* `>` – Returns True if left operand is greater than right.
* `<` – Returns True if left operand is less than right.
* `>=` – Returns True if left operand is greater than or equal to right.
* `<=` – Returns True if left operand is less than or equal to right.


```python
x1 = 10
x2 = 10
print(x1 == x2)
print(x1 != x2)
print(x1 > x2)
print(x1 < x2)
print(x1 >= x2)
print(x1 <= x2)
```

You can compare your variables with different numbers:

```python
print(x1 <= 12 <= x2)
print(x1 <= 15 and x2 >= 3)
```

<h3>4. Functions and methods</h3>


You can find all Python built-in functions here: https://docs.python.org/3/library/functions.html

Some functions work with different types of variables of data collections:

```python
v1 = 'this is string'
v2 = [101, 102, 103, 104, 105]
v3 = {'course 1': ['student 1', 'student 2'], 'course 2': ['student 3', 'student 4'],
     'course 3': ['student 5', 'student 6']}

print(len(v1)) # for string - number of symbols
print(len(v2)) # for lists - number of elements inside
print(len(v3)) # for dictionaries - number of pairs key-value
```

But others won't work the same way:

```python
v1 = [101, 102, 103, 104, 105]
v2 = 'some_text'

print(sum(v1))
print(sum(v2))
```

Also, you need to understand differences between **functions** and **methods**. A function exists independently and does not belong to any object. You can call it from another module or use built-in functions.

```python
print(sum)

import math

print(math.factorial)

print(math.factorial(3)) # 3 + 2 + 1 = 6 (factorial of 3 is 6)
```

A method is a function that is bound to an object:

```python
my_string = 'password is: qwerty'
print(my_string.replace('qwerty', '')) # we use .replace() method (method for strings)

my_list = ['password', 'is', 'qwerty']
print(my_list.replace('qwerty', '')) # we face the Attribute Error
```

When you work in Python, you will inevitably encounter various **exceptions**. They protect you from more serious errors and indicate what you did wrong. 

Here you can find the most widespread exceptions:

`NameError`: It occurs when you try to access a variable, function, or object that has not been defined.


```python
print(username)
```

`TypeError:`: It occurs when an operation or function is applied to an object of an inappropriate type.


```python
v1 = 5
v2 = 'cat'
print(v1 + v2)
```

`ValueError`: It occurs when an operation or function receives an object of the correct type but with an invalid value

```python
number = int('Hello World')
```

You need to understand how to handle each type of exceptions, and after some practice, you will be able to do this. You can find all Python exceptions here: https://www.w3schools.com/python/python_ref_exceptions.asp
